```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file from your Drive

df = pd.read_csv('/content/Customer_banking.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.head(11)
```

```python
# Check for missing values

df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 11'])
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns={'EducationLevel': 'Education'})
```

```python
df.describe()
```

```python
# Histograms
df['Age'].hist()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
# Box plots
sns.boxplot(x='Education', y='Income', data=df)
plt.xlabel('Education')
plt.ylabel('Income')
plt.title('Box Plot of Income by Education')
plt.show()
```

```python
# Scatter plots
plt.scatter(df['Age'], df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Income vs Age')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Income', hue='Education', data=df)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Scatter Plot of Age vs. Income by Education')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='CustomerSatisfaction', data=df)
plt.xlabel('Customer Satisfaction')
plt.ylabel('Count')
plt.title('Count Plot of Customer Satisfaction')
plt.show()
```

```python
sns.barplot(x='Education', y='Income', data=df)
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.title('Bar Chart of Average Income by Education')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
sns.countplot(x='Education', data=df)
plt.xlabel('Education Level')
plt.ylabel('Number of Customers')
plt.title('Count of Customers by Education')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.hist(df['Income'], bins=10)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Histogram of Income')
plt.show()
```
